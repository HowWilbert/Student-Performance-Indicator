### 1) Problem statement
- This project understands how the student's performance (test scores) is affected by other variables such as Gender, Ethnicity, Parental level of education, Lunch and Test preparation course.


### 2) Data Collection
- Dataset Source - https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?datasetId=74977

### 3.8 Adding columns for "Total Score" and "Average"

```python
df['Total_score'] = df['math_score'] + df['reading_score']+df['writing_score']
df['Average'] = df['Total_score']/3
df.head()
```

Now we will try to know how many students scored full marks in each subject.

```python
math_full = df[df['math_score']==100].shape[0]
reading_full = df[df['reading_score']==100].shape[0]
writing_full = df[df['writing_score']==100].shape[0]

print(f"{math_full} number of students scored full in Maths")
print(f"{reading_full} number of students scored full in Reading")
print(f"{writing_full} number of students scored full in Writing")
```

```python
reading_less_20 = df[df['reading_score'] <= 20].shape[0]
writing_less_20 = df[df['writing_score'] <= 20].shape[0]
math_less_20 = df[df['math_score'] <= 20].shape[0]

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Female','Male'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group_Average= df.groupby('gender')['Average'].mean()
gender_group_math = df.groupby('gender')['math_score'].mean()
gender_group_writing = df.groupby('gender')['writing_score'].mean()
gender_group_reading = df.groupby('gender')['reading_score'].mean()
```

```python
plt.figure(figsize=(15, 8))

X = ['Total Average','Math Average','Reading Average','Writing Average']


female_scores = [gender_group_Average['female'], gender_group_math['female'],gender_group_reading['female'],gender_group_writing['female']]
male_scores = [gender_group_Average['male'], gender_group_math['male'],gender_group_reading['male'],gender_group_writing['male']]

X_axis = np.arange(len(X))
  
plt.bar(X_axis -0.1, male_scores, 0.2, label = 'Male')
plt.bar(X_axis +0.1, female_scores, 0.2, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
